```python
import pandas as pd
```

```python
data = pd.read_csv('titanic_data.csv')
```

```python
data.columns
```

```python
data = data[['PassengerId', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp',
       'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked','Survived']]
```

```python
data = data.drop(['PassengerId','Name','Ticket','Cabin'], axis = 1)
```

```python
data.dropna(inplace = True)
```

```python
data.reset_index(inplace = True)
data = data.drop(['index'], axis = 1)
```

```python
target = data['Survived']
data.drop(['Survived'], axis = 1, inplace = True)
```

```python
data['Sex'][data['Sex'] == 'male'] = 1
data['Sex'][data['Sex'] == 'female'] = 2

data['Embarked'][data['Embarked'] == 'S'] = 1
data['Embarked'][data['Embarked'] == 'C'] = 2
data['Embarked'][data['Embarked'] == 'Q'] = 3
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier


model = RandomForestClassifier()
model.fit(X_train,y_train)

predicted_y = model.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score

score = roc_auc_score(y_test,predicted_y)
```
